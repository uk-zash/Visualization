# Categorical ScatterPlots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = sns.load_dataset("tips")
```

```python
data.sample(5)
```

```python
data["tip"].max()
```

```python
sns.catplot(x = "day" , y = "tip" , kind = "strip" , data = data)
```

```python
sns.catplot(x = "day" , y = "tip" , kind = "swarm" , data = data)
```

```python
sns.catplot(x = "day" , y = "tip", hue="smoker" , kind = "swarm" , data = data)
```
