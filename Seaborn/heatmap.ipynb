```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= sns.load_dataset("flights")
```

```python
data.sample(5)
```

```python
x = data.pivot_table(index = "year" , columns = "month" , values = "passengers" , aggfunc="sum")
```

```python
sns.heatmap(x , cbar = False , linewidths=0.5 , annot=True , fmt="d" , cmap="viridis")
```
